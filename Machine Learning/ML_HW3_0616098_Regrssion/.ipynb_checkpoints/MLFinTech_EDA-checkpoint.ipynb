```python
# get the rows
dataset = dataset.iloc[0:29999,:]
# drop some columns
dataset = dataset.drop(['ID'], axis = 1)
```

```python
xCategorical = ['default.payment.next.month', 'SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
for col in xCategorical:    
    print(col)
    xtemp = dataset[col]
    print(xtemp.value_counts(normalize=True)) # relative frequencies 
    print('\n')
```

```python
dataset['LIMIT_BAL'].hist()
```

```python
sns.distplot(dataset['BILL_AMT1'], hist = True)
```

```python
col_names = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6'] 
fig, ax = plt.subplots(len(col_names), figsize=(16,12))
for i, col_val in enumerate(col_names):
    sns.distplot(dataset[col_val], hist = True, ax = ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize = 8)
    ax[i].set_ylabel('Count', fontsize = 8)
    
plt.show()
```

```python
col_names = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'] 
fig, ax = plt.subplots(len(col_names), figsize=(16,12))
for i, col_val in enumerate(col_names):
    sns.distplot(dataset[col_val], hist = True, ax = ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize = 8)
    ax[i].set_ylabel('Count', fontsize = 8)
    
plt.show()
```

```python
x = dataset['BILL_AMT1']
y = dataset['BILL_AMT2']
plt.scatter(x, y)       
plt.xlabel('BILL_AMT1')
plt.ylabel('BILL_AMT2')
plt.show()
```
