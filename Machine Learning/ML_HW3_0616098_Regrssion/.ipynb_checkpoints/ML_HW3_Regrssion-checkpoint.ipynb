### import module

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

# Linear Regression

## Data Generation

```python
low = -1.5
high = 1
numbers = 1000
x = np.random.uniform(low, high, (numbers, 1))
# yi = 3 * xi ^ 3 + 2 * xi ^ 2 − 3 * xi + 1 + ϵi
y = 3 * x ** 3 + 2 * x ** 2 - 3 * x + 1 + np.random.normal(0, 0.5, (numbers, 1))
```

### plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='orange')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```

## Data Preprocessing

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
max_degree = 4
polynomial_features = []
polynomial_features_model = []
for K in range(1, max_degree + 1):
    print('K =', K)
    polynomial_features_model.append(PolynomialFeatures(K).fit(x))
    print(polynomial_features_model[-1].transform(x))
    polynomial_features.append(polynomial_features_model[-1].transform(x))
```

## Model Construction

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regr = LinearRegression(fit_intercept=True)
linear_regr_model = []
for idx, K in enumerate(range(1, max_degree + 1)):
    linear_regr_model.append(linear_regr.fit(polynomial_features[idx], y))
```

## Results 

```python
x_test = np.linspace(-1.5, 1, 100)
x_test = np.reshape(x_test, (-1, 1))

print('x test :', x_test.T)
for idx, K in enumerate(range(1, max_degree + 1)):
    print('K :', K)
    reg = linear_regr_model[idx].fit(polynomial_features[idx], y)
    print('weights:', reg.coef_)
    print('intercept:', reg.intercept_)
    equations = 'yi = ' + str(reg.intercept_[0])
    for i in range(len(reg.coef_[0])):
        if reg.coef_[0][i] > 0:
            equations += ' + ' + str(reg.coef_[0][i]) + ' * xi ^ ' + str(i)
        elif reg.coef_[0][i] < 0:
            equations += ' - ' + str(-reg.coef_[0][i]) + ' * xi ^ ' + str(i)
    print('equation:', equations)
    print('yhat:')
    y_hat = reg.predict(polynomial_features_model[idx].transform(x_test))
    print('', y_hat.T)
```

### figure 

```python
colors = ['green', 'blue', 'black', 'red']
lw = 2

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='orange')

for idx, K in enumerate(range(1, max_degree + 1)):
    reg = linear_regr_model[idx].fit(polynomial_features[idx], y)
    equations = str(np.round(reg.intercept_[0], 2))
    for i in range(len(reg.coef_[0])):
        if reg.coef_[0][i] > 0:
            equations += ' $+$ %.2f $\\times$' % (reg.coef_[0][i]) + f' $x_i^{i} $'
        elif reg.coef_[0][i] < 0:
            equations += ' $-$ %.2f $\\times$' % (-reg.coef_[0][i]) + f' $x_i^{i} $'
    y_hat = reg.predict(polynomial_features_model[idx].transform(x_test))
    plt.plot(x_test, y_hat, color=colors[idx], linewidth=lw, label=equations)

plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend(loc='upper right')

plt.show()
```

# Logistic Regression

## Data Generation

```python
total_num = 1000
cov = [[0.1, 0], [0, 0.1]]
x = []
y = np.random.randint(2, size=(total_num))
y0 = []
y1 = []
for i in range(total_num):
    mean = [y[i], y[i]]
    point = (np.random.multivariate_normal(mean, cov, 1))[0]
    x.append(point)
    if y[i] == 0:
        y0.append(point)
    else:
        y1.append(point)

x = np.array(x)
x0 = x[:, 0]
x1 = x[:, 1]

y0 = np.array(y0)
y0_x0 = y0[:, 0]
y0_x1 = y0[:, 1]

y1 = np.array(y1)
y1_x0 = y1[:, 0]
y1_x1 = y1[:, 1]
```

### plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(y0_x0, y0_x1, c='blue', label='$y_i$ = 0')
plt.scatter(y1_x0, y1_x1, c='orange', label='$y_i$ = 1')
plt.legend(loc='upper left', fontsize=18)
plt.xlabel(r'$X_{i0}$', fontsize=16)
plt.ylabel(r'$X_{i1}$', fontsize=16)
plt.show()
```

## Model Construction

```python
from sklearn.linear_model import LogisticRegression
```

### L2-norm as the objective function

```python
logistic_regr = LogisticRegression(penalty='l2').fit(x, y)
```

## Results 

### model accuracy

```python
accuracy = logistic_regr.score(x, y)
print('accuracy:', accuracy)
```

### model weights and the corresponded terms

```python
print('coefficients:', logistic_regr.coef_)
print('intercept:', logistic_regr.intercept_)

equations = 'yi = L(' + str(logistic_regr.intercept_[0]) + ' + ' + str(logistic_regr.coef_[0][0]) + ' * x_i0' + ' + ' + str(logistic_regr.coef_[0][1]) + ' * x_i1)' 
print('equations:', equations)
```

### the decision boundary with a figure

```python
# Retrieve the model parameters.
b = logistic_regr.intercept_[0]
w1, w2 = logistic_regr.coef_.T

# Calculate the intercept and gradient of the decision boundary.
intercept = -b / w2
slope = -w1 / w2

# Plot the data and the classification with the decision boundary.
xmin, xmax = -1.5, 2
ymin, ymax = -1, 2
xd = np.array([xmin, xmax])
yd = slope * xd + intercept
plt.figure(figsize=(10, 6))
plt.plot(xd, yd, 'k', lw=1, ls='--')
plt.fill_between(xd, ymin, yd, color='tab:blue', alpha=0.3)
plt.fill_between(xd, yd, ymax, color='tab:orange', alpha=0.3)

plt.scatter(*x[y == 0].T, label='$y_i$ = 0')
plt.scatter(*x[y == 1].T, label='$y_i$ = 1')
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.xlabel(r'$X_{i0}$', fontsize=16)
plt.ylabel(r'$X_{i1}$', fontsize=16)

plt.legend(loc='upper right', fontsize=18)
plt.show()
```
