### (b) Randomly assign a cluster label to each observation. Report the cluster labels for each observation.

```python
labels = [1] * int(np.floor(len(x1) / 2)) + [2] * int(np.ceil(len(x1) / 2))
random.shuffle(labels)
print("labels =", labels)
```

### (c) Compute the centroid for each cluster.

#### Method1

```python
X = list(zip(x1, x2))
clf = NearestCentroid()
clf.fit(X, labels)
print(clf.centroids_)
```

#### Method2

##### cast to numpy

```python
labels_np = np.array(labels)
x1_np = np.array(x1)
x2_np = np.array(x2)
```

```python
def get_centroid(x1_np, x2_np, labels_np):
    one_arr = np.ones(len(x1))
    one_idx = (labels_np == one_arr) 
    centroid1 = [np.mean(x1_np[one_idx]), np.mean(x2_np[one_idx])]
    # idx_1 = [idx for idx, value in enumerate(labels) if value == 2]
    centroid2 = [np.mean(x1_np[np.where(labels_np == 2)]), np.mean(x2_np[np.where(labels_np == 2)])]
    return centroid1, centroid2
```

```python
centroid1, centroid2 = get_centroid(x1_np, x2_np, labels_np)
print("centroid1 ", centroid1)
print("centroid2 ", centroid2)
```

#### plot

```python
# x1_with_centroid = x1 + [centroid1[0], centroid2[0]]
# x2_with_centroid = x2 + [centroid1[1], centroid2[1]]
# y_with_centroid = y + [1, 2]
# plt.scatter(x1_with_centroid, x2_with_centroid, c=y_with_centroid)
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2, c=labels)
plt.scatter([centroid1[0], centroid2[0]], [centroid1[1], centroid2[1]], c=[1, 2], marker=u'+', s=150)
```

### (d)  Assign each observation to the centroid to which it is closest, in terms of Euclidean distance. Report the cluster labels for each observation

```python
def euclid(x1, x2, centroid_x1, centroid_x2):
    return ((x1 - centroid_x1) ** 2 + (x2 - centroid_x2) ** 2) ** 0.5
```

```python
def assign_label(x1, x2, centroid1, centroid2):
    if euclid(x1, x2, centroid1[0], centroid1[1]) <= euclid(x1, x2, centroid2[0], centroid2[1]):
        return 1
    else:
        return 2
```

```python
for idx in range(len(labels)):
    labels[idx] = assign_label(x1[idx], x2[idx], centroid1, centroid2)
print("labels =", labels)
```

#### plot

```python
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2, c=labels)
plt.scatter([centroid1[0], centroid2[0]], [centroid1[1], centroid2[1]], c=[1, 2], marker=u'+', s=150)
```

### (e) Repeat (c) and (d) until the answers obtained stop changing.

```python
while 1:
    labels_np = np.array(labels)
    new_centroid1, new_centroid2 = get_centroid(x1_np, x2_np, labels_np)
    if new_centroid1 == centroid1 and new_centroid2 == centroid2:
        break
    else:
        centroid1, centroid2 = new_centroid1, new_centroid2
        print("new centroid1 ", centroid1)
        print("new centroid2 ", centroid2)
        new_labels = labels.copy()
        for idx in range(len(new_labels)):
            new_labels[idx] = assign_label(x1[idx], x2[idx], centroid1, centroid2)
        if new_labels == labels:
            break
        else:
            labels = new_labels
            print("new labels =", labels)
            plt.scatter(x1, x2, c=labels)
            plt.scatter([centroid1[0], centroid2[0]], [centroid1[1], centroid2[1]], c=[1, 2], marker=u'+', s=150)
            plt.xlabel('X1')
            plt.ylabel('X2')
            plt.title("change label")
            plt.show()
```

### (f) In your plot from (a), color the observations according to the cluster labels obtained.

#### plot

```python
plt.scatter(x1, x2, c=labels)
plt.scatter([centroid1[0], centroid2[0]], [centroid1[1], centroid2[1]], c=[1, 2], marker=u'+', s=150)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title("The last Result")
plt.show()
```

## Chapter 10: Exercise 4. Suppose that for a particular data set, we perform hierarchical clustering using single linkage and using complete linkage. We obtain two dendrograms.

### (a) At a certain point on the single linkage dendrogram, the clusters {1, 2, 3} and {4, 5} fuse. On the complete linkage dendrogram, the clusters {1, 2, 3} and {4, 5} also fuse at a certain point. Which fusion will occur higher on the tree, or will they fuse at the same height, or is there not enough information to tell?

Not enough information to tell. The maximal intercluster dissimilarity could be equal or not equal to the minimial intercluster dissimilarity. If the dissimilarities were equal, they would fuse at the same height. If they were not equal, they single linkage dendogram would fuse at a lower height.<br>
For example, if d(1, 4) = 2, d(1, 5) = 3, d(2, 4) = 1, d(2, 5) = 3, d(3, 4) = 4 and d(3, 5) = 1, the single linkage dissimilarity between {1, 2, 3} and {4, 5} would be equal to 1 and the complete linkage dissimilarity between {1, 2, 3} and {4, 5} would be equal to 4. Thus, with single linkage, they would fuse at a height of 1, and with complete linkage, they would fuse at a height of 4. However, if all inter-observations distance are equal to 2, we would have that the single and complete linkage dissimilarities between {1, 2, 3} and {4, 5} are equal to 2.

### (b) At a certain point on the single linkage dendrogram, the clusters {5} and {6} fuse. On the complete linkage dendrogram, the clusters {5} and {6} also fuse at a certain point. Which fusion will occur higher on the tree, or will they fuse at the same height, or is there not enough information to tell?

They would fuse at the same height because linkage does not affect leaf-to-leaf fusion.<br>
For example, if d(5, 6) = 2, the single and complete linkage dissimilarities between {5} and {6} would be equal to 2. Thus, they would fuse at a height of 2 for single and complete linkage.

### (a) Analysis the data by K-means clustering. Explain how you choose the number of cluster and illustrate the answer clearly.

### Method1： K-means

```python
from sklearn.cluster import KMeans
```

#### (1)Pre-processing

```python
# In order to draw 2-dim plot
sliced_dataset_Kmeans = dataset.iloc[:, 1:3]
print("sliced_dataset:", sliced_dataset_Kmeans, sep='\n')
```

```python
# change the type：dataframe --> array
sliced_dataset_Kmeans_np = sliced_dataset_Kmeans.values
print('sliced_dataset_Kmeans_np type: ', type(sliced_dataset_Kmeans_np))
```

#### (2) Set K points as cluster centers (means)

```python
SSE = []  
for k in range(1, 9):
    estimator = KMeans(n_clusters = k)  
    estimator.fit(sliced_dataset_Kmeans_np)
    SSE.append(estimator.inertia_)
X = range(1, 9)
plt.xlabel('The number of clustering')
plt.ylabel('SSE') # within-cluster variation
plt.plot(X, SSE, 'o-')
plt.show()
```

#### k選擇3，意即分3群，因為原始資料本來就是分成三群。而且由上圖可看出，k為3的SSE已經是k為1的不到七分之一了，而且k在3之後所減少的SSE並不大

```python
# set the k and run algorithm
k = 3
kmeans_fit = KMeans(n_clusters = k).fit(sliced_dataset_Kmeans_np)
```

#### (3) Print and visualize the clustering results

```python
y_pred = KMeans(n_clusters = k).fit_predict(sliced_dataset_Kmeans_np)
print('y_pred: ', y_pred)
plt.figure(figsize = (10, 6))
plt.xlabel(sliced_dataset_Kmeans.columns[0])
plt.ylabel(sliced_dataset_Kmeans.columns[1])
plt.scatter(sliced_dataset_Kmeans_np[:, 0], sliced_dataset_Kmeans_np[:, 1], c=y_pred) # argument c: color
plt.show()
```

#### (4) Print and visualize the original class(wheat_class) of raw data

```python
plt.figure(figsize = (10, 6))
plt.xlabel(sliced_dataset_Kmeans.columns[0])
plt.ylabel(sliced_dataset_Kmeans.columns[1])
plt.scatter(sliced_dataset_Kmeans_np[:, 0], sliced_dataset_Kmeans_np[:, 1], c=dataset.iloc[:, -1].values)
plt.show()
```

### Explain how you choose the number of cluster and illustrate the answer clearly.

* k選擇3，意即分3群，因為原始資料本來就是分成三群。而且由上圖可看出，k為3的SSE已經是k為1的不到七分之一了，而且k在3之後所減少的SSE並不大
* 由上兩張圖可知，clustering results和real class(wheat_class)算是非常接近，real class(wheat_class)有較多重疊的區域，而Kmeans通常區域的劃分會較乾脆，所以沒有重疊的部分，不管有沒有撇除重疊區域部分，兩個結果算是很接近的。而Kmean是找local而非global minimum，因此也許還不是最佳解，但從兩張圖得知已經是很接近了。

### (b) Analysis the data by the hierarchical clustering. Under which similarity metric and linkage method does it produce clusters closest to the original classes? 

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering   # Hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

#### (1) Pre-processing

```python
sliced_dataset_HC = dataset.drop(dataset.columns[-1], axis = 1)
print('sliced_dataset_HC', sliced_dataset_HC)
```

```python
# change the type：dataframe --> array
sliced_dataset_HC_np = sliced_dataset_HC.values
```

#### (2) Cluster the data by hierarchical clustering.

```python
hclust = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward').fit(sliced_dataset_HC_np)
```

#### (3) Print the result coming of hierarchical clustering.

```python
cluster_labels_hclust = hclust.labels_
print("Clustering Result：", cluster_labels_hclust, sep='\n')
```

```python
Affinity = ['euclidean', 'l1', 'l2', 'manhattan', 'cosine']
Linkage = ['ward', 'complete', 'average', 'single']

for a in Affinity:
    for l in Linkage:
        if l == 'ward' and a != 'euclidean':
            continue
        hclust = AgglomerativeClustering(n_clusters=3, affinity=a, linkage=l).fit(sliced_dataset_HC_np)
        cluster_labels_hclust = hclust.labels_
        print('affinity:', a, '\tlinkage:', l, "\tClustering Result：")
        print(cluster_labels_hclust)
```

由於原資料是三種類別數量相等(都是70筆)的情況，因此由上圖個種類得數量分布情形，可以認為**linkage method為single時或similarity metric為cosine的時候表現** ***很差*** **，其餘的結果會是大約三種數量均分，與原分類最為接近**

#### (4) Draw a hierarchical clustering diagram

```python
method = 'single'
linked = linkage(sliced_dataset_HC_np, method)
plt.figure(figsize=(20, 20))
dendrogram(linked, orientation = 'top', labels = sliced_dataset_HC.index, show_leaf_counts = True)
plt.show()
```

```python
methods = ['single', 'complete', 'average', 'weighted']
only_euclidean_methods = ['centroid', 'median', 'ward']
metrics=['euclidean', 'cosine', 'jaccard', 'minkowski', 'correlation']
for method in methods:
    for metrix in metrics:
        linked = linkage(sliced_dataset_HC_np, method, metric=metrix)    
        dendrogram(linked, orientation = 'top', labels = sliced_dataset_HC.index, show_leaf_counts = True)
        print('method:', method, '\tmetric:', metrix)   
        plt.show()
for method in only_euclidean_methods:
    linked = linkage(sliced_dataset_HC_np, method)    
    dendrogram(linked, orientation = 'top', labels = sliced_dataset_HC.index, show_leaf_counts = True)
    print('method:', method, '\tmetric:', 'euclidean')   
    plt.show()
```

### Under which similarity metric and linkage method does it produce clusters closest to the original classes?

由於原資料是三種類別數量相等(都是70筆)的情況，因此由上圖個種類得數量分布情形，可以認為**linkage method為complete時且similarity metric為euclidean或minkowski的時候**，結果會是大約三種數量均分，故認為與原分類最為接近
