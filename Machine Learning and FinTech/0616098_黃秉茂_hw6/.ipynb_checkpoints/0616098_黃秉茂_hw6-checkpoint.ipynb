## 2. Pre-process the continuous and discrete data.

### import module

```python
from sklearn.preprocessing import LabelEncoder
```

### observe data for processing

### Process the continuous data (Age, Siblings/Spouses Aboard, Parents/Children Aboard, and Fare).

```python
# 0 ~ max -> 0 ~ 1
dataset['Siblings/Spouses Aboard'] /= dataset['Siblings/Spouses Aboard'].max()
dataset['Parents/Children Aboard'] /= dataset['Parents/Children Aboard'].max()
# z-score standardlize
dataset['Age'] = (dataset['Age'] - dataset['Age'].mean()) / dataset['Age'].std()
dataset['Fare'] = (dataset['Fare'] - dataset['Fare'].mean()) / dataset['Fare'].std()
```

### Process the discrete data (Pclass and Survived).

Since all the discrete data are categorical, I decide not to process them.

### Process the categorical data (Sex) - change the category data into dummy variable(label encoder).

```python
# method1
labelencoder = LabelEncoder()
for categorical_variables_cols in categorical_variables_cols_list:
    dataset[categorical_variables_cols] = labelencoder.fit_transform(dataset[categorical_variables_cols]) + 1
# method2    
for categorical_variables_cols in categorical_variables_cols_list:
    dummies = pd.get_dummies(dataset.loc[:, categorical_variables_cols], prefix = categorical_variables_cols, drop_first = True ) 
    dataset[categorical_variables_cols] = dummies + 1
```

```python
def recall_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

def precision_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2 * ((precision * recall) / (precision + recall + K.epsilon()))
```

### neural network model

```python
# build NN model
model = models.Sequential()
# add layers
model.add(layers.Dense(units=16, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(units=16, activation='relu'))
model.add(layers.Dense(units=1, activation='sigmoid'))
# optimizer: 使用的演算法
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy', f1_m, recall_m])
 
# start training, fit model
history = model.fit(X_train, Y_train, epochs=20, batch_size=128)
# pridict with the model
Y_pred_NN = model.predict_classes(X_test)
# get confusion matric
confusion_matric_NN = confusion_matrix(Y_test, Y_pred_NN, labels=[1, 0])
confusion_matric_NN = confusion_matric_NN.T
# print
print('neural network Confusion Matrix:')
print(confusion_matric_NN)
```

### define function

```python
def Measures(confusion_matrix):
    
    TP, FP, FN, TN = confusion_matrix[0, 0], confusion_matrix[0, 1], confusion_matrix[1, 0], confusion_matrix[1, 1]

    # Sensitivity, hit rate, recall, or true positive rate
    TPR = TP / (TP + FN)
    # Specificity or true negative rate
    TNR = TN / (TN + FP) 
    # Precision or positive predictive value
    PPV = TP / (TP + FP)
    # Negative predictive value
    NPV = TN / (TN + FN)
    # Fall out or false positive rate
    FPR = FP / (FP + TN)
    # False negative rate
    FNR = FN / (TP + FN)
    # False discovery rate
    FDR = FP / (TP + FP)

    # Overall accuracy
    ACC = (TP + TN) / (TP + FP + FN + TN)
    
    # F1 score
    F1 = (2 * TP) / (2 * TP + FN + FP)
    
    return TPR, ACC, F1
```
